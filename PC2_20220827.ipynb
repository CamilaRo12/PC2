```python
import pandas as pd
#importar el archivo y poner como multiindex los títulos
archivo = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQ19QV9wZKYU25WXhM0FaGSt5Kjb-ilWJTGRAjRSqMY77EX2KQFyCy9mVBWo_4v4fkF7P2RC5MuLtz_/pub?gid=545366491&single=true&output=csv',header=[0,1,2,3])
archivo
```

```python
archivo.info() #verificando nombres y tipos
```

```python
#se eliminan las columnas que solo tienen letras (a e f etc)
archivo = archivo.drop(columns=[('Unnamed: 3_level_0', 'Unnamed: 3_level_1', 'Unnamed: 3_level_2', 'Unnamed: 3_level_3') ,
                                ('Unnamed: 5_level_0', 'Unnamed: 5_level_1', 'Unnamed: 5_level_2', 'Unnamed: 5_level_3'),
                                ('Unnamed: 7_level_0', 'Unnamed: 7_level_1', 'Unnamed: 7_level_2', 'Unnamed: 7_level_3'),
                                ('Unnamed: 9_level_0', 'Unnamed: 9_level_1', 'Unnamed: 9_level_2', 'Unnamed: 9_level_3'),
                                ('Unnamed: 11_level_0', 'Unnamed: 11_level_1', 'Unnamed: 11_level_2', 'Unnamed: 11_level_3'),
                                ('Unnamed: 13_level_0', 'Unnamed: 13_level_1', 'Unnamed: 13_level_2', 'Unnamed: 13_level_3')])
archivo
```

```python
archivo = archivo.drop(index=archivo.index[226:272]) #se eliminan las últimas filas de notas
archivo
```

```python
#se debe reemplaza comas por puntos porque python solo lee .
archivo = archivo.replace(',','.', regex=True)
archivo
```

```python
import numpy as np
#para los numéricos, se quitan los .. y - que hay en algunos datos
for column in archivo.select_dtypes(include=np.number).columns:
    archivo[column] = archivo[column].replace(r'\.\.|—', pd.NA, regex=True)
archivo
```

```python
archivo.info()
```

```python
print(archivo.dtypes)
```

```python
# pasar de multi a single index
archivo.columns = ['_'.join(col).strip() for col in archivo.columns.values]

# ahora se quita los espacios
archivo.columns = archivo.columns.str.replace('\s', '', regex=True)
```

```python
print(archivo.dtypes) #para ver el nuevo nombre de las columnas
```

```python
# Assuming 'column_name' is the name of the numeric column you want to convert
archivo['Table1.HumanDevelopmentIndexanditscomponents_Unnamed:1_level_1_Unnamed:1_level_2_Unnamed:1_level_3'] = archivo['Table1.HumanDevelopmentIndexanditscomponents_Unnamed:1_level_1_Unnamed:1_level_2_Unnamed:1_level_3'].astype(str)
```
